```python
import pandas as pd
import numpy as np
import tkinter as tk
from tkinter import messagebox

# Define the dataset
ds = [
    ['Sunny', 'Hot', 'High', 'Weak', 'No'],
    ['Sunny', 'Hot', 'High', 'Strong', 'No'],
    ['Overcast', 'Hot', 'High', 'Weak', 'Yes'],
    ['Rain', 'Mild', 'High', 'Weak', 'Yes'],
    ['Rain', 'Cool', 'Normal', 'Weak', 'Yes'],
    ['Rain', 'Cool', 'Normal', 'Strong', 'No'],
    ['Overcast', 'Cool', 'Normal', 'Strong', 'Yes'],
    ['Sunny', 'Mild', 'High', 'Weak', 'No'],
    ['Sunny', 'Cool', 'Normal', 'Weak', 'Yes'],
    ['Rain', 'Mild', 'Normal', 'Weak', 'Yes'],
    ['Sunny', 'Mild', 'Normal', 'Strong', 'Yes'],
    ['Overcast', 'Mild', 'High', 'Strong', 'Yes'],
    ['Overcast', 'Hot', 'Normal', 'Weak', 'Yes'],
    ['Rain', 'Mild', 'High', 'Strong', 'No'],
]

# Create a DataFrame
df = pd.DataFrame(ds, columns=['Outlook', 'Temperature', 'Humidity', 'Wind', 'Play'])

# Calculate conditional probabilities using crosstab
d = pd.crosstab(df['Outlook'], df['Play'])
e = pd.crosstab(df['Temperature'], df['Play'])
f = pd.crosstab(df['Humidity'], df['Play'])
g = pd.crosstab(df['Wind'], df['Play'])

p_yes = df['Play'].value_counts(normalize=True)['Yes']
p_no = df['Play'].value_counts(normalize=True)['No']

def predict(outlook, temperature, humidity, wind):
    p_yes_given_x = (d.loc[outlook, 'Yes'] / df['Play'].value_counts()['Yes']) * \
                   (e.loc[temperature, 'Yes'] / df['Play'].value_counts()['Yes']) * \
                   (f.loc[humidity, 'Yes'] / df['Play'].value_counts()['Yes']) * \
                   (g.loc[wind, 'Yes'] / df['Play'].value_counts()['Yes']) * p_yes
    p_no_given_x = (d.loc[outlook, 'No'] / df['Play'].value_counts()['No']) * \
                  (e.loc[temperature, 'No'] / df['Play'].value_counts()['No']) * \
                  (f.loc[humidity, 'No'] / df['Play'].value_counts()['No']) * \
                  (g.loc[wind, 'No'] / df['Play'].value_counts()['No']) * p_no
    if p_yes_given_x > p_no_given_x:
        return 'Yes'
    else:
        return 'No'

# Create a tkinter window
window = tk.Tk()
window.title("Weather Prediction")

# Create input labels and entry boxes
outlook_label = tk.Label(window, text="Outlook")
outlook_label.grid(row=0, column=0)
outlook_box = tk.Entry(window)
outlook_box.grid(row=0, column=1)

temp_label = tk.Label(window, text="Temperature")
temp_label.grid(row=1, column=0)
temp_box = tk.Entry(window)
temp_box.grid(row=1, column=1)

humid_label = tk.Label(window, text="Humidity")
humid_label.grid(row=2, column=0)
humid_box = tk.Entry(window)
humid_box.grid(row=2, column=1)

wind_label = tk.Label(window, text="Wind")
wind_label.grid(row=3, column=0)
wind_box = tk.Entry(window)
wind_box.grid(row=3, column=1)

# Create output label
output_label = tk.Label(window, text="")
output_label.grid(row=4, column=1)

# Create predict function
def predict_weather():
    outlook = outlook_box.get()
    temperature = temp_box.get()
    humidity = humid_box.get()
    wind = wind_box.get()
    result = ""
    try:
        result = predict(outlook, temperature, humidity, wind)
    except:
        messagebox.showerror("Error", "Invalid input")
    output_label.config(text=result)

# Create a prediction button
button = tk.Button(window, text="Predict", command=predict_weather)
button.grid(row=4, column=0)

# Run the tkinter event loop
window.mainloop()
```
